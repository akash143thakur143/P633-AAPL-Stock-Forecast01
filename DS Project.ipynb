### Time Series Forecasting

```python
pj['Date'] = pd.to_datetime(pj['Date'])
pj = pj.sort_values('Date')

series = pj.set_index('Date')['Adj Close']
series.head()
```

```python
split = int(len(series) * 0.8)

train = series.iloc[:split]
test  = series.iloc[split:]

print("Train size:", len(train))
print("Test size:", len(test))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

def mape(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

def evaluate_model(name, y_true, y_pred):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae  = mean_absolute_error(y_true, y_pred)
    mp   = mape(y_true, y_pred)
    
    print(f"--- {name} ---")
    print("RMSE:", rmse)
    print("MAE :", mae)
    print("MAPE:", mp)
    return rmse, mae, mp
```

```python
from statsmodels.tsa.arima.model import ARIMA

arima_model = ARIMA(train, order=(5,1,0))
arima_fit = arima_model.fit()

arima_pred = arima_fit.forecast(steps=len(test))

evaluate_model("ARIMA", test, arima_pred)
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

sarima_model = SARIMAX(train, order=(1,1,1), seasonal_order=(1,1,1,12))
sarima_fit = sarima_model.fit(disp=False)

sarima_pred = sarima_fit.forecast(steps=len(test))

evaluate_model("SARIMA", test, sarima_pred)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

train_scaled = scaler.fit_transform(train.values.reshape(-1,1))
test_scaled  = scaler.transform(test.values.reshape(-1,1))
```

```python
def create_sequences(data, window=60):
    X, y = [], []
    for i in range(window, len(data)):
        X.append(data[i-window:i])
        y.append(data[i])
    return np.array(X), np.array(y)

window = 60

X_train, y_train = create_sequences(train_scaled, window)
X_test, y_test   = create_sequences(test_scaled, window)

print(X_train.shape, y_train.shape)
```

```python
!pip install tensorflow
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], 1)),
    LSTM(50),
    Dense(1)
])

model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=1)
```

```python
lstm_pred_scaled = model.predict(X_test)

lstm_pred = scaler.inverse_transform(lstm_pred_scaled)
y_test_real = scaler.inverse_transform(y_test)

evaluate_model("LSTM", y_test_real, lstm_pred)
```

```python
results = []

rmse, mae, mp = evaluate_model("ARIMA", test, arima_pred)
results.append(["ARIMA", rmse, mae, mp])

rmse, mae, mp = evaluate_model("SARIMA", test, sarima_pred)
results.append(["SARIMA", rmse, mae, mp])

rmse, mae, mp = evaluate_model("LSTM", y_test_real, lstm_pred)
results.append(["LSTM", rmse, mae, mp])

comparison = pd.DataFrame(results, columns=["Model", "RMSE", "MAE", "MAPE"])
comparison
```

####  Report 

Time-based Split:

The dataset was split into training and testing sets using chronological order to avoid data leakage. The first 80% of the time series was used for training and the remaining 20% was used for testing.

Models:

ARIMA and SARIMA were used as statistical forecasting models, while LSTM was used as a deep learning model for time-series prediction.

Evaluation:

Models were evaluated using RMSE, MAE, and MAPE. Lower values indicate better forecasting performance.
